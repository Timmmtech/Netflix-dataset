### Q.2 In which year highest number of the TV Shows & Movies were released? Show with Bar Graph

```python
data.dtypes
```

# 
Since we'll have to work with date column, we'll have to convert the object data type to datetime

```python
data['Date_N'] = pd.to_datetime(data['Release_Date'])
```

```python
data.head()
```

```python
data['Date_N'].dt.year.value_counts()      ## Highest number of tv shows & movies were released in 2019 
```

### Using  Bar graph to visualize the result

```python
highest = data['Date_N'].dt.year.value_counts()
highest
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
```

```python
highest.plot(kind='bar', color='Blue')
plt.xticks(rotation=75)
plt.title('Years highest no. of TV shows & Movies were released')
plt.xlabel("Years")
plt.ylabel("Number of TV Shows & Movies released");
```

### Q. 9 What is the max. duration of a Movie/Show on Netflix?

```python
data.Duration.dtypes
```

# 

This shows the Duration is in the object format. we will have to convert it to numeric format to work on it 

```python
data[['Minutes', 'Unit']] = data['Duration'].str.split(' ', expand=True)
```

### Q. 10 Which individual country has the Highest No. of Tv shows?

```python
data_tvshow = data[data['Category'] == 'TV Show']
data_tvshow.head(2)
```

```python
individual_country =data_tvshow.Country.value_counts().head(10)           ## Country with the highest No. of TV Shows
individual_country 
```

```python

sns.barplot(individual_country.index, individual_country)
plt.xticks(rotation=75)
plt.title('Individual Country with the highest No. of TV Show')
plt.xlabel("Country")
plt.ylabel(None);
```
